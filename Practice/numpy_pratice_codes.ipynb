What is the difference between arr[-1] and arr[:-1]?

arr[-1] returns the last value
arr[:-1] returns the all values except the last value

```python
import numpy as np

arr = np.array([10,20,20,30,40,40,50])
copyOfArr = arr.copy()
viewOfArr = arr.view()
replaceNum = np.where(arr>25,100,0)
any_greater_45 = np.any(arr>45)
all_greater_5 = np.all(arr>5)
remove_duplicate = np.unique(arr)
sort_arr = np.sort(arr)
index_sort_arr = np.argsort(arr)
b=np.array([60,70,80])
add_arr = np.concatenate((arr,b))
a=np.array([1,2,3])
b=np.array([4,5,6])
vertically_stack = np.vstack((a,b))
horizontally_stack = np.hstack((a,b))
arr2 = np.arange(1,11)
split_arr = np.split(arr2,2)

print(arr)
print(copyOfArr)
print(viewOfArr)
print(replaceNum)
print(any_greater_45)
print(all_greater_5)
print(remove_duplicate)
print(sort_arr)
print(index_sort_arr)
print(add_arr)
print(vertically_stack)
print(horizontally_stack)
print(split_arr)
```
